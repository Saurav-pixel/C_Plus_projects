1. Load the wine dataset from the given ‘wine.csv’ file into a variable (say data). 
2. From the variable, extract the numerical information into a new variable, say scores. 


```
import pandas as pd
import numpy as np
from scipy import stats
import statistics

##### question 1 ########
df= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/wine.csv',sep=',',header=None)
data =df.values
print(data)

###### question 2 ######
score1 = np.array(data)
#print(score1)

score = np.delete(score1,(0),axis=0)
score = score.astype(np.float)

print("\n")
# print("after removing from the first row from the matrix")
print(score)
print("\n")



```

3. Calculate the row-wise,column-wise, and overall mean, median, mode, and standard deviation of the dataset. In this question write separate functions for each operation and pass the variable(scores) as the parameter to the function. 


```
# finding mean, median,mode and standard deviation row-wise and column wise and overall .
import numpy as np
from scipy import stats

# overall mean
print("mean of matrix \n",np.mean(score))
print("\n")

# overall median 
print("median of matrix \n",np.median(score))
print("\n")

# overall mode of the .csv numeriacal value
print("overall mode \n",stats.mode(score))
print("\n")

# overall standard deviation
print("overall standard deviation is \n",np.std(score))

# row wise operation
print("mean of matrix row wise\n",score.mean(axis=1))
print()
print("median of matrix row-wise \n",np.median(score,axis=1))
print()
print("mode of matrix row-wise\n")
print(stats.mode(score,axis=1))


#column wise
print("mean of matrix column wise\n",score.mean(axis=0))
print()
print("median of matrix column-wise \n",np.median(score,axis=0))
print()
print("mode of matrix column-wise\n")
print(stats.mode(score,axis=0))
print("\n")
```
